```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
X_train.shape
```

```python
import numpy as np
import random
class VectorizedLinearRegression_GradientDescent:
    def __init__(self,lr=0.01,n_epoch=1000):
        self.lr=lr
        self.n_epoch=n_epoch
        self.W = np.random.rand(X.shape[1],1) - 0.5
        self.d = random.random() - 0.5
    def loss(self,X,Y):
        N = len(X) # number of samples
        y_hat = np.dot(X,self.W) + self.d
        f = y_hat - Y
        loss = np.sum(f**2)/N
        return loss
    def predict(self,X):
        y_hat = np.dot(X,self.W) + self.d
        return y_hat
    def fit(self, X,Y):
        N = len(X) # number of samples
        for i in range(self.n_epoch):
            y_hat = np.dot(X,self.W) + self.d
            f = y_hat - Y
            loss = np.sum(f**2)/N
            # Updating m and b
            self.W -= self.lr * (2 * X.T.dot(f) / N)
            self.d -= self.lr * (2 * f.sum()/N)
            print("Epoch",i+1,",loss",loss)
    
```

```python
model = VectorizedLinearRegression_GradientDescent(lr=0.001,n_epoch=20000)
model.fit(X_train, np.array(y_train).reshape(-1,1))
```
