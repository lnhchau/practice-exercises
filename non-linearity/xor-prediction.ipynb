```python
import numpy as np
import matplotlib.pyplot as plt
import random
data = np.genfromtxt('xor.csv',dtype=float,delimiter=',',skip_header=1)
data
```

```python
X = data[:,:-1]
Y = data[:,-1]
# Tách dữ liệu thành hai phần dựa trên đầu ra (0 và 1)
X1_0 = [X[i,0] for i in range(len(Y)) if Y[i] == 0]
X2_0 = [X[i,1] for i in range(len(Y)) if Y[i] == 0]
X1_1 = [X[i,0] for i in range(len(Y)) if Y[i] == 1]
X2_1 = [X[i,1] for i in range(len(Y)) if Y[i] == 1]

# Vẽ dữ liệu
plt.scatter(X1_0, X2_0, c='blue', marker='o', label='y=0')
plt.scatter(X1_1, X2_1, c='red', marker='x', label='y=1')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='best')
plt.title('XOR Data')
plt.grid(True)
plt.show()
```

```python
import numpy as np   
import random
class VectorizedLinearRegression_GradientDescent:
    def __init__(self,lr=0.01,n_epoch=1000):
        self.lr=lr
        self.n_epoch=n_epoch
        self.W = np.random.rand(2,1) - 0.5
        self.d = random.random() - 0.5
    def fit(self, X,Y):
        N = len(X) # number of samples
        for i in range(self.n_epoch):
            y_hat = np.dot(X,self.W) + self.d
            f = y_hat - Y
            loss = np.sum(f**2)/N
            # Updating m and b
            self.W -= self.lr * (2 * X.T.dot(f) / N)
            self.d -= self.lr * (2 * f.sum()/N)

            print("Epoch",i+1,",loss",loss)
```

```python
model = VectorizedLinearRegression_GradientDescent(lr=0.0001)
model.fit(X, np.array(Y).reshape(-1,1))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

y_hat = np.dot(X,model.W) + model.d
# Vẽ dữ liệu
ax.scatter(X[:,0], X[:,1], y_hat, c=y_hat, cmap='viridis', marker='o')
X1_plane, X2_plane = np.meshgrid(np.linspace(0, 1, 10), np.linspace(0, 1, 10))
Z_plane = X1_plane + X2_plane

# Plot the plane as a wireframe
ax.plot_wireframe(X1_plane, X2_plane, Z_plane, color='b', alpha=0.5)

# Set labels for the axes
ax.set_xlabel('Input 1')
ax.set_ylabel('Input 2')
ax.set_zlabel('Output')

# Set the title for the 3D plot
ax.set_title('XOR Data and Plane in 3D')

plt.show()
```
